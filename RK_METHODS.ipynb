# **Solving the first order ordinary differential equation using Runge-Kutte method**

Let us consider that we have a first order differential equation of form $\dfrac{dy}{dx}= f(x,y)$ given $y(x_0)=y_0$.

Method of solving

a) Euler's method : It introduces an error of order $\mathcal{O}(h^2)$

## **Formula for the first iteration (i.e; from $y_0= y(x_0)$ to $y_1=y(x_0+h)$) of the Euler's method**

$y_1= y_0+k_1$

where,

$k_1= h*f(x_0,y_0)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function
def f(x,y):
  r= x+y
  return r

# Define a Function for Euler's method' 
def Euler(x0,y0,xn,n):

    h = (xn-x0)/n   # step size h
    x=[x0]          # list of x and y to append data
    y=[y0]
    
    for i in range(1,n):
      k1 = h*f(x0, y0)
      y1 = y0 + k1
      x1= x0 + h

      x.append(x1)
      y.append(y1)

      y0= y1         # assign the modified values og x0 and y0 for the next iteration
      x0= x1

    plt.xlabel('x') # Label of 'x' axis
    plt.ylabel('y') # Label of 'y' axis
    plt.style.use('seaborn-whitegrid')
    plt.title('Title') # Title of plot
    plt.plot(x,y)    # Print the solution

    return  
```

```python
# Inputs initial conditions

x0 = 0 
y0 = 1
xn = 4
n=1000

Euler(x0,y0,xn,n)
```

## **RK2 Method: Formula for first order ODE**

It introduces an error of $\mathcal{O}(h^3)$. Conveince yourself 

$y_1= y_0 + c_1k_1+c_2k_2$

There are many versions of RK2 depending on the value of $c_1, c_2, k_1 \& k_2$. We will be working with a specific form of RK2 where

$c_1=c_2= 1/2$

$k_1 = h* f(x_0,y_0)$

$k_2= h* f(x_0+h, y_0+k_1)$

# $$k_2=h*f(x_0+\frac{h}{2},y_0+\frac{k_1}{2})$$

Use the code given above (for Euler's method) and make appropriate change in the program to apply RK2 algorithm.


```python
# write code for RK2 here
import numpy as np
import matplotlib.pyplot as plt

# Define a function
def f(x,y):
  r= x+y
  return r

# Define a Function for Euler's method' 
def rk2(x0,y0,xn,n,c1,c2):

    h = (xn-x0)/n   # step size h
    x=[x0]          # list of x and y to append data
    y=[y0]
    
    for i in range(1,n):
      k1 = h*f(x0, y0)
      k2 = h*f(x0+(h/2),y0+(k1/2))
#       y0+=c1*k1+c2*k2
#       x0+=h
      y1=y0+c1*k1+c2*k2
      x1=x0+h
        
      x.append(x1)
      y.append(y1)
    
      y0= y1         # assign the modified values og x0 and y0 for the next iteration
      x0= x1

    plt.xlabel('x') # Label of 'x' axis
    plt.ylabel('y') # Label of 'y' axis
    plt.style.use('seaborn-whitegrid')
    plt.title('Title') # Title of plot
    plt.plot(x,y)    # Print the solution

    return  
```

## **RK4 Method: Formula for first order ODE**

It introduces an error of $\mathcal{O}(h^5)$. Conveince yourself arguably

$y_1= y_0 + c_1k_1+c_2k_2+c_3 k_3+c_4k_4$

There are many versions of RK2 depending on the value of $c_1, c_2, k_1 \& k_2$. We will be working with a specific form of RK2 where

$c_1=1/6, c_2= 1/3. c_3= 1/3, c_4= 1/6$

$y_1= \dfrac{1}{6}(k1+2k_2+2k_3+k_4)$

where

$k_1 = h* f(x_0,y_0)$

$k_2= h* f(x_0+h/2, y_0+k_1/2)$

$k_3= h* f(x_0+h/2, y_0+k_2/2)$

$k_4= h* f(x_0+h, y_0+k_3)$



Use the code given above (for Euler's method) and make appropriate change in the program to apply RK4 algorithm.

```python
# write code for RK4 here
import numpy as np
import matplotlib.pyplot as plt

# Define a function
def f(x,y):
  r= x+y
  return r

# Define a Function for Euler's method' 
def rk4(x0,y0,xn,n,c1,c2,c3,c4):

    h = (xn-x0)/n   # step size h
    x=[x0]          # list of x and y to append data
    y=[y0]
    
    for i in range(1,n):
      k1 = h*f(x0, y0)
      k2 = h*f(x0+(h)/2,y0+(k1)/2)
      k3=h*f(x0+(h)/2,y0+(k2)/2)
      k4=h*f(x0+h,y0+k3)
      y1=y0+c1*k1+c2*k2+c3*k3+c4*k4
      x1=x0+h
        
      x.append(x1)
      y.append(y1)
    
      y0= y1         # assign the modified values of x0 and y0 for the next iteration
      x0= x1

    plt.xlabel('x') # Label of 'x' axis
    plt.ylabel('y') # Label of 'y' axis
    plt.style.use('seaborn-whitegrid')
    plt.title('Title') # Title of plot
    plt.plot(x,y)    # Print the solution

    return  
```

# **Solving the Second order  differential equation using Runge-Kutte methods**

Let us consider that we have a differential equation to solve $\dfrac{d^2y}{dt^2}= f(y,t)$ with initial value conditions (IVC) $y(t_0)=y_0$  and $y'(t_0)=y'_0 $ 

A second order differential equation can always be written as a system of two coupled first order differential equation i.e;

$\dfrac{dy}{dt}= x$    .................   $      y(t_0)=y_0$

$\dfrac{dx}{dt}= f(x,y,t)$ ............... $     x(t_0)=y'_0$


In a more general sense, a system of two simultaneous coupled first order differential equation is eventually a second order ODE. So we can write down a  general  form as 

$\dfrac{dx}{dt}= f(x,y,t) $

$\dfrac{dy}{dt}= g(x,y,t)$

Siilar to first order ODE we can use the Euler's method, RK2 and RK4 method to solve the system.

Again, I will provide you with the code for Euler's method and you have to modify it for RK2 AND RK4 methods



##  **Euler's method for second order ODE**

Let usstart with a system of first order ODE

$\dfrac{dx}{dt}= f(x,y,t) $

$\dfrac{dy}{dt}= g(x,y,t)$


having IVC $x(t_0)=x_0$ and $y(t_0)=y_0$

We will solve the system as follows

$x_1= x_0+k_1$

$y_1= y_0+l_1$

where, $x_1= x(t=t_0+h)$ and $y_1= y(t=t_0+h)$

$k_1= h*f(x_0,y_0,t_0)$

$l_1= h*g(x_0,y_0,t_0)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions
def f(x,y,t):
  r= y
  return r

def g(x,y,t):
  r= x+t
  return r

# Define a Function for Euler's method' 
def Euler2(x0,y0,t0,tn,n):

    h = (tn-t0)/n   # step size h
    t=[t0]          # list of t, x and y to append data
    x=[x0]
    y=[y0]
    
    for i in range(0,n):
      k1 = h*f(x0, y0, t0)
      l1 = h*g(x0, y0, t0)

#       x1 = x0 + k1
#       y1 = y0 + l1
#       t1 = t0 + h
      x0 +=  k1
      y0 +=  l1
      t0 +=  h

      x.append(x0)
      y.append(y0)
      t.append(t0)

#       x0 = x1         # assign the modified values og x0 and y0 for the next iteration
#       y0 = y1
#       t0 = t1

    return (x,y,t)
```

```python
# Inputs initial conditions

x0 = 0 
y0 = 0
t0 = 0
tn = 4
n=1000
x, y, t =Euler2(x0,y0,t0,tn,n);
```

```python
plt.xlabel('t') # Label of 'x' axis
plt.ylabel('x') # Label of 'y' axis
plt.style.use('seaborn-whitegrid')
plt.title('Title') # Title of plot
plt.plot(t,x)    # Print the solution
plt.grid()
plt.show()
```

## **RK2 Method: Formula for coupled first order ODE**

It introduces an error of $\mathcal{O}(h^3)$. Conveince yourself 


$x_1= x_0 + \dfrac{1}{2}(k_1+k_2)$

$y_1= y_0 + \dfrac{1}{2}(l_1+l_2)$

where

$k_1 = h* f(x_0,y_0,t_0)$

$l_1 = h* g(x_0,y_0,t_0)$

$k_2= h* f(x_0+k_1, y_0+l_1, t_0+h)$

$l_2= h* g(x_0+k_1, y_0+l_1, t_0+h)$


Use the code given above (for Euler's method) and make appropriate change in the program to apply RK2 algorithm.

```python
# Code for RK2 HERE
import numpy as np
import matplotlib.pyplot as plt

# Define functions
def f(x,y,t):
  r= y
  return r

def g(x,y,t):
  r= x+t
  return r

# Define a Function for Euler's method' 
def rk2(x0,y0,t0,tn,n):

    h = (tn-t0)/n   # step size h
    t=[t0]          # list of t, x and y to append data
    x=[x0]
    y=[y0]
    
    for i in range(1,n):
      k1 = h*f(x0, y0, t0)
      l1 = h*g(x0, y0, t0)
      k2=h*f(x0+k1/2,y0+l1/2,t0+h/2)
      l2=h*g(x0+k1/2,y0+l1/2,t0+h/2)
        
      x0 += 0.5*(k1+k2)
      y0 += 0.5*(l1+l2)
      t0 += h

      x.append(x0)
      y.append(y0)
      t.append(t0)

#       x0 = x1         # assign the modified values og x0 and y0 for the next iteration
#       y0 = y1
#       t0 = t1

    return (x,y,t)
```

$x_1= x_0 + \dfrac{1}{6}(k_1+2k_2+2k_3+k_4)$

$l_3= h* g(x_0+k_2/2, y_0+l_2/2, t_0+h/2)$

$k_4= h* f(x_0+k_3, y_0+l_3, t_0+h)$

$l_4= h* g(x_0+k_3, y_0+l_3, t_0+h)$
