```python
import math as m
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
def innerproduct(u,v):
    y = u*v
    gfg = integrate.simps(y,X)         
    return gfg 
def Norm(v):
    return ((innerproduct(v,v))**(1/2))
def unitvector(v):
    return v/Norm(v)
```

```python
def series(v,U):     
    add=0
    for i in range(0,len(U)):
        add=add+(innerproduct(v,U[i]))*U[i]
    return add
```

```python
def Gram_schmidt(V):      ## V is linearly independent set
    n=len(V)              ## n is number of linear independent vectors
    u1=unitvector(V[0])
    U=[u1]
    for i in range(1,n):
        U2=V[i]-series(V[i],U)        ## U2 is orthogonal vector
        u2=unitvector(U2)             ## u2 is orthonormal vector 
        U.append(u2)
    return U          ## U is orthonormal set
```

```python
n=3                 ## Degree of polynomial less than equal to n
X=np.linspace(-1,1,100)       
V=[]                              ## V is vector space
for i in range(0,n+1):
    z=X**i
    V.append(z)
```

```python
U=Gram_schmidt(V)     ## U is a orthonormal vector space

for i in range(0,n+1):
    plt.plot(X,U[i])
```

```python
def f(x):
    return m.exp(x)
Y=[]             
for i in X:
    Y.append(f(i))     ## Y is vector with its element as value of function on points    

Z=0
for i in range(0,n+1):
    Z=Z+(innerproduct(Y,U[i]))*U[i]       ## Z is vector with elements as values calculated using polynomial
#print(Z)
plt.plot(X,Z,"r")      ## Approximated using polynomial
plt.plot(X,Y,'g')      ## origional function
```
