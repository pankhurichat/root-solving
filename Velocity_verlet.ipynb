```python
#string SHM, RK4
import numpy as np
import matplotlib.pyplot as plt
import math as m
# Define functions

def f(x,v,t):
  r= v
  return r

def g(x,v,t):
  a=-x
  return a

# Define a Function for Euler's method' 
def rk4(x0,y0,t0,tn,n):

    h = (tn-t0)/n   # step size h
    t=[t0]          # list of t, x and y to append data
    x=[x0]
    y=[y0]
    
    for i in range(1,n):
      k1 = h*f(x0, y0, t0)
      l1 = h*g(x0, y0, t0)
      k2=h*f(x0+(k1)/2,y0+(l1)/2,t0+(h)/2)
      l2=h*g(x0+(k1)/2,y0+(l1)/2,t0+(h)/2)
      k3=h*f(x0+(k2)/2,y0+(l2)/2,(h)/2)
      l3=h*g(x0+(k2)/2,y0+(l2)/2,(h)/2)
      k4=h*f(x0+k3,y0+l3,t0+h)
      l4=h*g(x0+k3,y0+l3,t0+h)
        
      x1 = x0 + (1/6)*(k1+2*k2+2*k3+k4)
      y1 = y0 + (1/6)*(l1+2*l2+2*l3+l4)
      t1 = t0 + h

      x.append(x1)
      y.append(y1)
      t.append(t1)

      x0 = x1         # assign the modified values og x0 and y0 for the next iteration
      y0 = y1
      t0 = t1

    return (x,y,t)

# def Euler2(x0,y0,t0,tn,n):

#     h = (tn-t0)/n   # step size h
#     t=[t0]          # list of t, x and y to append data
#     x=[x0]
#     y=[y0]
    
#     for i in range(1,n):
#       k1 = h*f(x0, y0, t0)
#       l1 = h*g(x0, y0, t0)

#       x1 = x0 + k1
#       y1 = y0 + l1
#       t1 = t0 + h

#       x.append(x1)
#       y.append(y1)
#       t.append(t1)

#       x0 = x1         # assign the modified values og x0 and y0 for the next iteration
#       y0 = y1
#       t0 = t1

#     return (x,y,t)

# def rk2(x0,y0,t0,tn,n):

#     h = (tn-t0)/n   # step size h
#     t=[t0]          # list of t, x and y to append data
#     x=[x0]
#     y=[y0]
    
#     for i in range(1,n):
#       k1 = h*f(x0, y0, t0)
#       l1 = h*g(x0, y0, t0)
#       k2=h*f(x0+k1,y0+l1,t0+h)
#       l2=h*g(x0+k1,y0+l1,t0+h)
        
#       x1 = x0 + 0.5*(k1+k2)
#       y1 = y0 + 0.5*(l1+l2)
#       t1 = t0 + h

#       x.append(x1)
#       y.append(y1)
#       t.append(t1)

#       x0 = x1         # assign the modified values og x0 and y0 for the next iteration
#       y0 = y1
#       t0 = t1

#     return (x,y,t)
```

```python
#verlet
def ver(x0,v0,a0,t0,tn):
    h = 10**(-3)
    n=int((tn-t0)/h)
    t=[t0]          
    x=[x0]
    v=[v0]
    a=[a0]
    
    for i in range(0,n):
    
        
      x1 = x0 + h*v0+((h**2)/2)*a0
    
      a1=g(x1,(t0+h))
    
      v1 = v0 + (h/2)*(a1+a0)
        
      t1 = t0+h
    
      x.append(x1)
      v.append(v1)
      a.append(a1)
      t.append(t1)

      x0 = x1         
      v0 = v1
      t0 = t1
      a0 = a1

    return x,v,t
```
